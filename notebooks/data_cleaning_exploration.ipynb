Tokenization & Stopwords Removal

```python
import re
from nltk.corpus import stopwords

#stopwords
stop_words = set(stopwords.words('english'))

#Keep important words
stop_words -= {"not", "no", "i"}

#Clean text
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)        # remove URLs
    text = re.sub(r"@\w+", "", text)           # remove mentions
    text = re.sub(r"#(\w+)", r"\1", text)      # keep hashtag word
    text = re.sub(r"[^a-z\s]", "", text)       # remove punctuation/symbols
    return text

tweet_df_copy = tweet_df[['text', 'emotion']].copy()

tweet_df_copy['clean_text'] = tweet_df_copy['text'].apply(clean_text)

# create token
tweet_df_copy['token'] = tweet_df_copy['clean_text'].str.split()

##Remove stopwords (and short tokens)
tweet_df_copy['cleaned_token'] = tweet_df_copy['token'].apply(
    lambda tokens: [
        word for word in tokens 
        if word not in stop_words and len(word) > 1
    ]
)

## explode the token list
tweet_df_explode = tweet_df_copy.explode('cleaned_token')

##Drop empty tokens

tweet_df_explode = tweet_df_explode[tweet_df_explode['cleaned_token'].notna()]

## Group df by emotion
tweet_df_grouped = tweet_df_explode.groupby(['emotion'])

tweet_df_grouped.head(5)
```

```python
# Show all rows and columns
pd.set_option("display.max_rows", None)     # Show all rows
pd.set_option("display.max_columns", None)  # Show all columns
pd.set_option("display.width", None)        # Prevent line wrapping
pd.set_option("display.max_colwidth", None) # Show full column content

##Word frequency by emotion

analysis_df = tweet_df_explode[['emotion', 'cleaned_token']].copy()

tweet_count = analysis_df.groupby(['emotion','cleaned_token']).size().sort_values(ascending= False).reset_index(name='count')

#Pivot so tokens are rows and emotions are columns
tweet_pivot = tweet_count.pivot(index='cleaned_token', columns='emotion', values='count').fillna(0)

#sort rows by total counts across emotions
tweet_pivot['total'] = tweet_pivot.sum(axis=1)
tweet_pivot = tweet_pivot.sort_values(by='total', ascending=False).drop(columns='total')

tweet_pivot.head(10)
```

Bi-gram Analysis

```python
from nltk.util import bigrams

tweet_df_copy['bigram'] = tweet_df_copy['cleaned_token'].apply(
    lambda tokens: [" ".join(pair) for pair in bigrams(tokens)]
)

bigram_df = tweet_df_copy[['emotion', 'bigram']].explode('bigram')

bigram_df = bigram_df[bigram_df['bigram'].notna()]

bigram_df.head(10)
```

```python
## Count bigrams per emotion

bigram_count = (
    bigram_df
    .groupby(['emotion', 'bigram'])
    .size()
    .sort_values(ascending=False)
    .reset_index(name='count')
)
```

```python
## Pivot to show bigram count per emotion

bigram_pivot = bigram_count.pivot(
    index='bigram',
    columns='emotion',
    values='count'
).fillna(0)

#sort rows by total counts across emotions
bigram_pivot['total'] = bigram_pivot.sum(axis=1)
bigram_pivot = bigram_pivot.sort_values(by='total', ascending=False).drop(columns='total')

bigram_pivot.head(10)
```

```python
## Bar chart to show top words per emotion

def plot_top_words_by_emotion(df, top_n=20):
    cols = df.columns
    n_cols = 2
    n_rows = math.ceil(len(cols) / n_cols)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 5 * n_rows))
    axes = axes.flatten()

    color_map = {"anger": "red",
                "joy": "gold",
                "sadness": "black",
                "optimism": "green"}

    for i, col in enumerate(cols):
        top_words = df[col].sort_values(ascending=False).head(top_n)

        #reverse for horizontal plot
        top_words = top_words.sort_values(ascending=True)
        
        ax = axes[i]
        ax.barh(top_words.index, top_words.values, color=color_map.get(col, "gray"))
        ax.set_title(f"{col}")
        ax.set_xlabel("Count")
        ax.set_ylabel("Words")

    plt.tight_layout()    
    plt.show()
```

Visualizing TF-IDF

```python
## Apply TFIDF

X = cleaned_df['clean_text']
y = cleaned_df['emotion']

le = LabelEncoder()

y_enc = le.fit_transform(y)

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(X)

#Apply PCA (dimension reduction method)

pca = PCA(n_components = 2)
pca_result = pca.fit_transform(tfidf_matrix.toarray())

#Apply t-SNE (dimension reduction method)
tsne = TSNE(n_components=2, perplexity=2, random_state=5)
tsne_result = tsne.fit_transform(tfidf_matrix.toarray())

##Apply UMAP
reducer = umap.UMAP(n_components=2, metric="cosine", random_state=42)
umap_embedding = reducer.fit_transform(tfidf_matrix) 
```

```python
idx = np.random.choice(len(X), size=30, replace=False)

#Plot TF-IDF 2D Graph - PCA

plt.figure(figsize=(10, 7))
scatter = plt.scatter(pca_result[:, 0], pca_result[:, 1], c = y_enc, edgecolor='k', s=50, alpha=0.7)

for i in idx:
    plt.annotate(X[i][:100], (pca_result[i, 0], pca_result[i, 1]), textcoords="offset points", xytext=(5,7),
                 fontsize = 8)

handles, _ = scatter.legend_elements()
plt.legend(
    handles,
    le.classes_,
    title="Emotion"
)
plt.title('PCA of TF-IDF Matrix')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
idx = np.random.choice(len(X), size=10, replace=False)

#Plot TF-IDF 2D Graph - tsne

plt.figure(figsize=(10, 7))
scatter = plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c = y_enc, edgecolor='k', s=50, alpha=0.7)

for i in idx:
    plt.annotate(X[i][:100], (tsne_result[i, 0], tsne_result[i, 1]), textcoords="offset points", xytext=(5,7),
                 fontsize = 8)

handles, _ = scatter.legend_elements()
plt.legend(
    handles,
    le.classes_,
    title="Emotion"
)
plt.title('tsne of TF-IDF Matrix')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

The 2D plot of the TF-IDF matrix provides a visual representation of the documents in the reduced dimensional space. Each point represents a document, and the distance between points indicates the similarity between the documents based on their TF-IDF scores. Closer points suggest similar content, while distant points indicate dissimilar content

```python
## Using plotly

tsne_df = pd.DataFrame({
    "C1": tsne_result[:, 0],
    "C2": tsne_result[:, 1],
    "text": X,
    "emotion": y
})

fig = px.scatter(
    tsne_df,
    x="C1",
    y="C2",
    color="emotion",
    hover_data={
        "text": True,
        "C1": False,
        "C2": False
    },
    title="Interactive tsne Plot of TF-IDF Features"
)

fig.show()
```

```python
## Using UMAP dimension reduction method

umap_df = pd.DataFrame({
    "C1": umap_embedding[:, 0],
    "C2": umap_embedding[:, 1],
    "text": X,
    "emotion": y
})

fig = px.scatter(
    umap_df,
    x="C1",
    y="C2",
    color="emotion",
    hover_data={
        "text": True,
        "C1": False,
        "C2": False
    },
    title="Interactive umap Plot of TF-IDF Features"
)

fig.show()
```

Insights

The visualization shows a high degree of overlap between emotion classes, indicating that TF-IDF features do not produce clearly separable clusters. While some dense regions exist, these reflect shared vocabulary rather than distinct emotional groupings. This suggests that emotion classification in this dataset is inherently complex and may require more context-aware representations
